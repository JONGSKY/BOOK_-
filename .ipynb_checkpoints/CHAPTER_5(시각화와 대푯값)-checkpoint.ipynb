```python
import pandas as pd
from matplotlib import pyplot as plt
import sklearn.datasets
```

#### 5.2 아이리스 데이터셋

```python
def get_iris_df():
    ds = sklearn.datasets.load_iris()
    df = pd.DataFrame(ds['data'], columns=ds['feature_names'])
    code_species_map = dict(zip(range(3), ds['target_names']))
    df['species'] = [code_species_map[c] for c in ds['target']]
    return df
```

```python
df = get_iris_df()
```

#### 5.3 원형 차트

```python
import matplotlib.font_manager as fm

font_path = 'C:/Windows/Fonts/H2GTRE.ttf'
fontprop = fm.FontProperties(fname=font_path, size=18)
# Font에 있는 글꼴 가져오기 -> 한글 폰트 사용 가능
```

```python
sums_by_species = df.groupby('species').sum()
var = 'sepal width (cm)'
sums_by_species[var].plot(kind='pie', fontsize=20)
plt.ylabel(var, horizontalalignment='left')
plt.title(var + '로 분류한 붓꽃', fontsize=25, fontproperties=fontprop)
plt.savefig('iris_pie_for_one_variable.png')
plt.close()
```

```python
sums_by_species = df.groupby('species').sum()
sums_by_species.plot(kind='pie', subplots=True,
                    layout=(2,2), legend=False)
plt.title('종에 따른 전체 측정값', fontproperties=fontprop)
plt.savefig('iris_pie_for_each_variable.png')
plt.close()
```

#### 5.4 막대그래프

```python
sums_by_species = df.groupby('species').sum()
var = 'sepal width (cm)'
sums_by_species[var].plot(kind='bar', fontsize=15, rot=30)

plt.title(var + '로 분류한 붓꽃', fontsize=20, fontproperties=fontprop)
plt.savefig('iris_bar_for_one_variable.png')
plt.close()

sums_by_species = df.groupby('species').sum()
sums_by_species.plot(kind='bar', subplots=True, fontsize=12)

plt.suptitle('종에 따른 전체 측정값', fontproperties=fontprop)
plt.savefig('iris_bar_for_each_variable.png')
plt.close()
```

```python
# 'fontsize'로 글자크기를 조절함
# 'rot'으로 글자를 회전함
# 'suptitle()' 메서드로 그림 전체 제목을 설정합니다. 판다스는 데이터프레임의 열 이름으로 그림 소제목을 자동으로 설정함
```

#### 5.5 히스토그램

```python
df.plot(kind='hist', subplots=True, layout=(2,2))
plt.suptitle('붓꽃 히스토그램', fontsize=20, fontproperties=fontprop)
plt.show()
```

```python
for spec in df['species'].unique():
    forspec = df[df['species']==spec]
    forspec['petal length (cm)'].plot(kind='hist', alpha=0.4, label=spec)
    
plt.legend(loc='upper right')
plt.suptitle('종에 따른 꽃잎 길이', fontproperties=fontprop)
# plt.show()
plt.savefig('iris_hist_by_spec.png')
```

#### 5.6 평균, 표준편차, 중간값, 백분위

```python
col = df['petal length (cm)']
average = col.mean()
std = col.std()
median = col.quantile(0.5)
percentile25 = col.quantile(0.25)
percentile75 = col.quantile(0.75)
```

```python
col = df['petal length (cm)']
perc25 = col.quantile(0.25)
perc75 = col.quantile(0.75)
clean_avg = col[(col > perc25) & (col < perc75)].mean()
```

#### 5.7 상자그림

```python
col = 'sepal length (cm)'
df['ind'] = pd.Series(df.index).apply(lambda i : i%50)
df.pivot('ind', 'species')[col].plot(kind='box')
plt.show()
```

#### 5.8 산포도

```python
df.plot(kind="scatter",
       x="sepal length (cm)", y="sepal width (cm)")
plt.title("길이 대 너비", fontproperties=fontprop)
plt.show()
```

```python
plt.close()
colors = ["r", "g", "b"]
markers = [".", "*", "^"]
fig, ax = plt.subplots(1,1)
for i, spec in enumerate(df['species'].unique()):
    ddf = df[df['species'] == spec]
    ddf.plot(kind="scatter",
            x="sepal width (cm)", y="sepal length (cm)",
            alpha=0.5, s=10*(i+1), ax=ax,
            color=colors[i], marker=markers[i], label=spec)
    
plt.legend()
plt.show()
```

#### 5.9 산포도와 로그 축

```python
import pandas as pd
import sklearn.datasets as ds
import matplotlib.pyplot as plt
```

```python
# 팬다스 데이터프레임 생성
bs = ds.load_boston()
df = pd.DataFrame(bs.data, columns=bs.feature_names)
df['MEDV'] = bs.target
```

```python
# 일반적인 산포도
df.plot(x='CRIM', y='ZN', kind='scatter')
plt.title('일반 축에 나타낸 범죄 발생률', fontproperties=fontprop)
plt.show()
```

```python
df.plot(x='CRIM', y='ZN', kind='scatter', logx=True)
plt.title('일반 축에 나타낸 범죄 발생률', fontproperties=fontprop)
plt.show()
```
